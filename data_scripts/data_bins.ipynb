```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import math

np.random.seed(0)
```

```python
X_res=np.genfromtxt('../final_datasets/X_training_100K.txt', delimiter=",")
Y= np.genfromtxt('../final_datasets/Y_training_100K.txt', delimiter=None)
print X_res.shape, Y.shape
```

```python
col_wise_X= np.transpose(X_res)
num_bins=10
print col_wise_X.shape
for i in xrange(len(X_res[0])):
    min_val=min(col_wise_X[i])
    max_val=max(col_wise_X[i])
    bins=[min_val]
    data=sorted(zip(col_wise_X[i],Y), key=lambda x: x[0])
    d0=[]
    d1=[]
    for j in xrange(1,len(data)+1):
        if int(data[j-1][1])==0:
            d0.append(data[j-1][0])
        else:
            d1.append(data[j-1][0])
        if j%(len(X_res)/num_bins) is 0:
            bins.append(data[j-1][0])
    print bins
    print len(d0), len(d1)
    print "Graph for Feature: ", i+1
    plt.hist([d0,d1],bins, histtype='bar', color=['b','g'])
    plt.show()
```
