# PCA

```python
import numpy as np
X_res=np.genfromtxt('../final_datasets/X_training_100K.txt', delimiter=",")
X_val=np.genfromtxt('../final_datasets/X_validation_50K.txt', delimiter=",")
X_tes=np.genfromtxt('../final_datasets/X_test_50K.txt', delimiter=",")
X_sub=np.genfromtxt('../final_datasets/X_submission_200K.txt', delimiter=",")
Y= np.genfromtxt('../final_datasets/Y_training_100K.txt', delimiter=None)
Y_tes= np.genfromtxt('../final_datasets/Y_test_50K.txt', delimiter=None)
Y_val= np.genfromtxt('../final_datasets/Y_validation_50K.txt', delimiter=None)
print X_res.shape, Y.shape, X_val.shape, Y_val.shape, X_tes.shape, Y_tes.shape, X_sub.shape
```

```python
from sklearn.preprocessing import PolynomialFeatures 
poly= PolynomialFeatures(2)

xtr_p = poly.fit_transform(X_res)
xte_p = poly.fit_transform(X_tes)
xva_p = poly.fit_transform(X_val)
xsu_p = poly.fit_transform(X_sub)
print xtr_p.shape, xte_p.shape, xva_p.shape, xsu_p.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(svd_solver='full')
xtr_pc= pca.fit_transform(xtr_p)
xte_pc= pca.fit_transform(xte_p)
xva_pc= pca.fit_transform(xva_p)
xsu_pc= pca.fit_transform(xsu_p)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(xtr_pc.n_components_),xtr_pc.explained_variance_)
plt.show()
```

```python
print xtr_pc.explained_variance_[:50]
```

#### Clearly we can see after 10 features the percentage of variance captured is less then ~5%. Thus picking first 10 features

```python
num_comp=10
xtr=xtr_pc[:,:num_comp]
xte=xte_pc[:,:num_comp]
xva=xva_pc[:,:num_comp]
xsu=xsu_pc[:,:num_comp]
print xtr.shape, xte.shape, xva.shape, xsu.shape
```

# SVM

```python
import matplotlib.pyplot as plt
def plotAUC(plot_name, x_tuple, *y_tuples):
    colors=['b','g','r','c','m','k']
    legends=[]
    for i in xrange(len(y_tuples)):
        a,=plt.plot(x_tuple[0],y_tuples[i][0],'-',c=colors[i],label=y_tuples[i][1])
        legends.append(a)
    plt.legend(handles=legends)
    plt.xlabel(x_tuple[1])
    plt.ylabel("AUC Values")
    plt.title(plot_name)
    plt.grid()
    plt.show()
```

```python
from sklearn.svm import SVC
learner = SVC(C=1.0, kernel='rbf', gamma='auto',probability=False, tol=0.001)
learner.fit(xtr[:10000],Y[:10000])
yh_val=learner.predict(xva[:5000])
```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(Y_val[:5000], yh_val, pos_label=1)
print metrics.auc(fpr, tpr)
```

```python
C=[0.1, 0.5, 1, 2, 3, 5, 7] 
tr_auc=[]
val_auc=[]
for c in C: 
    learner = SVC(C=c, kernel='rbf', gamma='auto',probability=False, tol=0.001)
    learner.fit(xtr[:10000],Y[:10000])
    yh_val=learner.predict(xva[:5000])
    fpr, tpr, thresholds = metrics.roc_curve(Y_val[:5000], yh_val, pos_label=1)
    val_auc.append(metrics.auc(fpr, tpr))
    yh=learner.predict(xtr[:10000])
    fpr, tpr, thresholds = metrics.roc_curve(Y[:10000], yh, pos_label=1)
    tr_auc.append(metrics.auc(fpr, tpr))
    print c ,'done'
print tr_auc
print val_auc
plotAUC('SVM c value variation', (C,'c_values'), (tr_auc, 'Training auc'), (val_auc, 'Validation auc'))
```

```python
gamma=[0.001, 0.004, 0.007, 0.01,0.013,0.016, 0.019,0.022]
tr_auc=[]
val_auc=[]
for g in gamma: 
    learner = SVC(C=5, kernel='rbf', gamma=g,probability=False, tol=0.001)
    learner.fit(xtr[:10000],Y[:10000])
    yh_val=learner.predict(xva[:5000])
    fpr, tpr, thresholds = metrics.roc_curve(Y_val[:5000], yh_val, pos_label=1)
    val_auc.append(metrics.auc(fpr, tpr))
    yh=learner.predict(xtr[:10000])
    fpr, tpr, thresholds = metrics.roc_curve(Y[:10000], yh, pos_label=1)
    tr_auc.append(metrics.auc(fpr, tpr))
    print g ,'done'
print tr_auc
print val_auc
plotAUC('SVM c value variation', (gamma,'gamma values'), (tr_auc, 'Training auc'), (val_auc, 'Validation auc'))
```

```python
from sklearn.ensemble import RandomForestClassifier
learner= RandomForestClassifier(criterion='gini',max_depth=25, max_features='auto', max_leaf_nodes=None,
                        min_samples_leaf=1, min_samples_split=2, n_estimators=10)
learner.fit(xtr_p, Y)
yh_val=learner.predict(xva_p)
fpr, tpr, thresholds = metrics.roc_curve(Y_val, yh_val, pos_label=1)
print metrics.auc(fpr, tpr)
```
